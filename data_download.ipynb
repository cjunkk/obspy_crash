```python
import obspy
from obspy.clients.fdsn import Client

client = Client("INGV")
```

```python
inventory = client.get_stations(network='IV', level="channel", 
                                starttime=obspy.UTCDateTime())
```

```python
print(inventory)
```

```python
inventory.write('./INGV/stazioni.xml',format='stationxml')
```
